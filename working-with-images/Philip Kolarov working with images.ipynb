```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage.io import imread
import cv2
```

```python
hb_img = imread('materials/hamburger.jpg')
```

```python
red = hb_img[:, :, 0]
green = hb_img[:, :, 1]
blue = hb_img[:, :, 2]
red_mean_br = np.mean(red)
green_mean_br = np.mean(green)
blue_mean_br = np.mean(blue)
red_mean_br, green_mean_br, blue_mean_br
```

```python
kernel = np.ones((3, 3), np.uint8)
opened_image = cv2.morphologyEx(blue, cv2.MORPH_OPEN, kernel)

plt.imshow(opened_image)
plt.show()
```

```python
white_pix_total = np.sum (opened_image == 241)
white_pix_total
```
