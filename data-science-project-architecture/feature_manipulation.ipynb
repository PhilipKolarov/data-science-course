**Hypothesis A**: There is no evident direct connection between one single variable and the probability of a patient obtaining a positive diagnosis

```python
for col in diagnosis_df.columns:
    f, ax = plt.subplots(1, 2)
    ax[0].hist(df[col], color='r', bins=25)
    ax[0].set_title(f'general {col} distribution')
    ax[0].set_xlabel('values')
    ax[0].set_ylabel('counts')
    ax[1].hist(diagnosis_df[col], color='r', bins=25)
    ax[1].set_title(f'diagnosis {col} distribution')
    ax[1].set_xlabel('values')
    ax[1].set_ylabel('counts')

    f.subplots_adjust(wspace=0.25, hspace=0.5)
    f.set_figwidth(12)
    f.set_figheight(4)
    plt.show()
```
