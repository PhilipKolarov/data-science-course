```python
df = pd.read_csv('data/asthma_disease_data.csv')
df
```

```python
df.columns
```

```python
df.dtypes
```

```python
plt.hist(df['Age'], bins=len(df['Age'].unique()))
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('No. of Patients')
plt.grid()
plt.show()
```

```python
plt.bar(df['Gender'].unique().tolist(), df['Gender'].value_counts())
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('No. of Patients')
plt.xticks(np.arange(0, 2, 1))
plt.show()
```

```python
plt.bar(df['Ethnicity'].unique().tolist(), df['Ethnicity'].value_counts())
plt.title('Ethnicity Distribution')
plt.xlabel('Ethnicity')
plt.ylabel('No. of Patients')
plt.xticks(np.arange(0, 4, 1))
plt.show()
```
